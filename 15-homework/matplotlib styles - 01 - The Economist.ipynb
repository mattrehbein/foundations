## rcParams

rcParams are settings for matplotlib. They're how stylesheets work (the thing when you use `plt.style.use`.

You can find examples of the different styles [here](https://matplotlib.org/2.1.1/gallery/style_sheets/style_sheets_reference.html) and [here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html).

> **A big long list is:** bmh, classic, dark_background, fivethirtyeight, ggplot, grayscale, seaborn-bright, seaborn-colorblind, seaborn-dark, seaborn-dark-palette, seaborn-darkgrid, seaborn-deep, seaborn-muted, seaborn-notebook, seaborn-paper, seaborn-pastel, seaborn-poster, seaborn-talk, seaborn-ticks, seaborn-white, seaborn-whitegrid 

If you think one of them is cool, you can find its rcparams styles by checking out its mplstyle file in [in this directory](https://github.com/matplotlib/matplotlib/tree/master/lib/matplotlib/mpl-data/stylelib).

# Bar graph (Single color)

```python
df = pd.read_csv("countries.csv")
result = df.groupby("continent").life_expectancy.median().sort_values()
result
```

```python
color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0]
ax = result.plot(kind='barh', figsize=(10, 5), color=color)

# CUSTOMIZATION CODE GOES UNDER HERE
ax.grid(False, axis='x')
# getting rid os unnecessary y axis label
ax.yaxis.label.set_visible(False)

# SAME AS ABOVE, LEAVING TITLE OUT BC IT TAKES AWAY THE BACKGROUND COLORS I WORKED HARD TO 
# MAKE LOOK LIKE THE ECONOMIST!!
# ax.set_title("Continental Life Expectancy", fontweight='bold', loc='left', pad=20)
# ax.title.set_fontsize(25)
# ax.title.set_fontname("Andale Mono")

# then we save it
plt.savefig("graph-bar.pdf")
plt.show()
```

# Line graph

```python
df = pd.read_csv("visas-by-continent-long.csv")
df = df.groupby(["year", "continent"]).visas.sum().unstack().reset_index()
df.drop(columns='Unknown', inplace=True)
df['year'] = pd.to_datetime(df.year, format="%Y")
df.set_index('year', inplace=True)
percent_df = df.div(df.sum(axis=1), axis='rows') * 100

percent_df.head(2)
```

```python


# CUSTOMIZATION CODE GOES UNDER HERE
plt.rcdefaults()
plt.rcParams.update({
  'patch.linewidth': 0.5,
  'patch.facecolor': 'white',
  'patch.edgecolor': '#EEEEEE',
  'patch.antialiased': True,
  'font.size': 10.0,
  'font.family': 'Andale Mono', # looks closest to 'Officina Sans'
  'axes.facecolor': 'white',
  'axes.edgecolor': 'white',
  'axes.linewidth': '1',
  'axes.grid': True,
  'grid.linewidth': '2',
  'legend.numpoints': '2',
  'axes.titlesize': 'x-large',
  'axes.labelsize': 'large',
  'axes.labelcolor': '#555555',
  'axes.axisbelow': True,
  'axes.prop_cycle': cycler('color', ['#7ad2f6', '#014d64', '#6794a7', '#ffd200', '#e37e00', '#00887d']),
  'xtick.major.bottom': True,
  'xtick.minor.bottom': False,
  'xtick.major.size': 7,
  'xtick.major.pad': 7,
  'xtick.direction': 'in',
  'xtick.labelsize': 'large',
  'ytick.left': False,
  'ytick.labelsize': 'large',
  'grid.color': 'lightgray',
  'grid.linestyle': '-',
  'figure.facecolor': 'white'
})

# HOW TO MOVE y axis LABELS FROM LEFT SIDE OF GRAPH TO RIGHT???

ax = percent_df.plot(kind='line', figsize=(10, 5), linewidth=4)
ax.grid(False, axis='x')
ax.xaxis.label.set_visible(False)
ax.yaxis.label.set_visible(False)

# then we save it
plt.savefig("graph-lines.pdf")
plt.show()
```

# Scatter plot

```python
df = pd.read_csv("countries.csv")
df.head(2)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Don't touch this part of the code!
colors = list(reversed(plt.rcParams['axes.prop_cycle'].by_key()['color']))

groups = df.groupby('continent')
for name, group in groups:
    group.plot(kind='scatter', x='gdp_per_capita', y='life_expectancy', color=colors.pop(), label=name, s= 5 + (group.population / 200000), ax=ax, alpha=0.75)
ax.legend_.remove()

# CUSTOMIZATION CODE GOES UNDER HERE
plt.rcdefaults()
plt.rcParams.update({
  'patch.linewidth': 0.5,
  'patch.facecolor': 'white',
  'patch.edgecolor': '#EEEEEE',
  'patch.antialiased': True,
  'font.size': 10.0,
  'font.family': 'Andale Mono', # looks closest to 'Officina Sans'
  'axes.facecolor': 'white',
  'axes.edgecolor': 'white',
  'axes.linewidth': '1',
  'axes.grid': True,
  'grid.linewidth': '0.5',
  'legend.scatterpoints': '2',
  'axes.titlesize': 'x-large',
  'axes.labelsize': 'large',
  'axes.labelcolor': '#555555',
  'axes.axisbelow': True,
  'axes.prop_cycle': cycler('color', ['#7ad2f6', '#014d64', '#C10534', '#ffd200', '#e37e00', '#00887d']),
  'xtick.major.bottom': True,
  'xtick.minor.bottom': False,
  'xtick.major.size': 7,
  'xtick.major.pad': 2,
  'xtick.direction': 'in',
  'xtick.labelsize': 'medium',
  'ytick.left': False,
  'ytick.labelsize': 'medium',
  'grid.color': '#d9e6eb',
  'grid.linestyle': '-',
  'figure.facecolor': 'white'
})

ax.grid(False, axis='x')
ax.xaxis.label.set_visible(True)
ax.yaxis.label.set_visible(True)

# then we save it
plt.savefig("graph-scatter.pdf")
plt.show()
```
